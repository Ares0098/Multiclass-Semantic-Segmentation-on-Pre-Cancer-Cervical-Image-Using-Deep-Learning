```python
def label_maker(point, label, shapes) :
    point_list = []
    label_list = point
    temp = np.zeros(shape=shapes)
    
    for item in label_list :
        if item['label'] == label :
            point_list.append(item['points'])

    for point in point_list :
        temp = cv2.fillPoly(temp, pts=np.int32([point]), color=(255, 255, 255))
        
    return temp 
```

```python
for index, label in enumerate(classes) :
    print("-----------------------------------------")
    print(f'{label} image creation process started !!')
    print("-----------------------------------------")
    for json_index, json_item in enumerate(json_list) :
        json_paths = json_path+json_item
        with open(json_paths) as json_file:
            data = json.load(json_file)
        img_name = data['imagePath']
        split = os.path.splitext(img_name)
        point = data['shapes']
        ada = img_name in img_list
    
        if (ada == True) :
            img_path = image_path+img_name
            img_point = point
            img = cv2.imread(img_path)
            img_shape = img.shape
            img_jadi = label_maker(img_point, classes_labelme[index], img_shape)
            filepath = label_path + label + '/' + split[0] + '.png'
            cv2.imwrite(filepath, img_jadi)
            print(f"Image '{split[0]}' with label '{label}' created!! ")
        
        else :
            print("------------------------------------------")
            print(f"Image {img_name} tidak ditemukan")
            print("Label tidak dapat dibuat !!")
            print("------------------------------------------")
            continue
                       
            
    print(f'{label} image creation process finished !!')
    print('===========================================')
    
    
```
