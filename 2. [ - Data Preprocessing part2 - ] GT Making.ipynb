## Real GT

```python
print("Ground Truth creation started")
print("===============================")

for idx, name in enumerate(img_name) :
    
    ca = cv2.imread(path_ca + name)
    clr = cv2.imread(path_clr + name)
    lesi = cv2.imread(path_lesi + name)
    tz = cv2.imread(path_tz + name)
    
    shape = ca.shape
    temp = np.zeros(shape=shape)
    for x in range(shape[0]) :
        for y in range(shape[1]) :
            
            a1 = sum(ca[x, y, :])
            a2 = sum(tz[x, y, :])
            a3 = sum(clr[x, y, :])
            a4 = sum(lesi[x, y, :])
            
            if ((a1 == 765) & (a2 == 0) & (a3 == 0) & (a4 == 0)) :
                # Perubahan warna untuk CA
                temp[x, y, :] = [1, 1, 1] 
                
            elif ((a1 == 765) & (a2 == 765) & (a3 == 0) & (a4 == 0)) :
                # Perubahan warna untuk TZ
                temp[x, y, :] = [2, 2, 2] 
                
            elif ((a1 == 765) & (a2 == 765) & (a3 == 765) & (a4 == 0)) :
                # Perubahan warna untuk CLR
                temp[x, y, :] = [3, 3, 3] 
                
            elif ((a1 == 765) & (a2 == 765) & (a3 == 0) & (a4 == 765)) :
                # Perubahan warna untuk LESI
                temp[x, y, :] = [4, 4, 4] 
                
            elif ((a1 == 765) & (a2 == 765) & (a3 == 765) & (a4 == 765)) :
                # Perubahan warna untuk LESI 2
                temp[x, y, :] = [4, 4, 4] 
                
            else :
                # Perubahan warna untuk Background
                temp[x, y, :] = [0, 0, 0]
    
    gtname = GT_path+file_name[idx]+".png"
    cv2.imwrite(gtname, temp)
    print(f"{gtname} Ground Truth Created!!")
            
```
